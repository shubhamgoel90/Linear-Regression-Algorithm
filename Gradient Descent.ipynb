# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
print(x)
```

```python
y = (x-5)**2
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

# Gradient Descent Algorithm

```python
x1 = np.random.randint(10)
print(x1)
```

```python
steps = 1
x1 =0                    # Random Variable
learning_rate = 0.1
plt.plot(x,y)
while(steps<=15):
    x1 = x1 -learning_rate*2*(x1-5)
    
    error = ((x1-5)**2)
    plt.scatter(x1,error)
    print("X %.4f , Error %.4f"%(x1,error))
    steps +=1
plt.show()
```
